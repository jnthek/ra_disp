```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_original = 1024
t_original = np.linspace(0,1,N_original)
sig_base = np.sin(2*np.pi*10*t_original) + np.sin(2*np.pi*20*t_original) + np.sin(2*np.pi*50*t_original)

sig_original_rfft = np.fft.rfft(sig_base)
sig_original = np.fft.irfft(sig_original_rfft) 
sig_original = sig_original * len(sig_original) # To normalise the amplitudes

interp_fac = 8

sig_interp_rfft = np.append(sig_original_rfft, np.zeros((interp_fac-1)*int(N_original/2))) #For rfft, only half of the spectrum is usefull
sig_interp = np.fft.irfft(sig_interp_rfft)
sig_interp = sig_interp * len(sig_interp) # To normalise the amplitudes

t_interp = np.linspace(0,1,interp_fac*N_original)
```

```python
plt.figure(figsize=(16,9))
plt.stem(t_original, sig_original, label="Original "+str(N_original)+" samples", color='black')
plt.stem(t_interp, sig_interp, label="Interpolated "+str(N_original*interp_fac)+" samples", color='red')
plt.xlim(0,0.05)
plt.xlabel("Time (s)")
plt.legend(loc='best')
```
